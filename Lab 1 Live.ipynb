```python
#Ex. 8 a)
reset()
t = var('t')
s = var('s')
f(t,s) = 1 - (s*cos(4*t) * cos(t))/sqrt(1-(s^2)*(cos(4*t))^2*sin(t)^2)
x(t) = f(t-pi/2, s=0.5)
y(t) = f(t, s=0.5)
parametric_plot(((x(t), y(t))), (t,0,2*pi))
```

```python
#Ex. 8 b)

x(t) = f(t-pi/2, 1/10)
y(t) = f(t, 1/10)
g=parametric_plot(((x(t), y(t))), (t,0,2*pi))
    
for k in [2..10]:
    x(t) = f(t-pi/2, k/10)
    y(t) = f(t, k/10)
    g1=parametric_plot(((x(t), y(t))), (t,0,2*pi))
    g=g+g1

show(g)    
```
