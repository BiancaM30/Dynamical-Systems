```python
#Ex 1 a)
reset()
x=var('x')
y=function('y')(x)
eqd=diff(y,x)==2*x*(1+y*y)
show(eqd)
```

```python
show(desolve(eqd, y))
```

```python
#daca vrem sa vedem ce metoda foloseste sage
desolve(eqd,y,show_method=True)
```

```python
#facem substitutia _C
sol = desolve(eqd, y)
_C=var('C')
sol.substitute(_C==1)
```

```python
[sola] = solve(sol, y(x))
f(x,_C) = sola.rhs()
f(x, _C)
```

```python
plot(f(x,1/2), x, -pi/2, pi/2, ymin=-10, ymax=10)
```
