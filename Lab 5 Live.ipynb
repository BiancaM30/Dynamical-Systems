```python
#Ex. 1
#a)
t=var('t')
w=var('w')
w0=var('w0')
assume(w0,'integer')
assume(w0>0)

x=function('x')(t)

eqd=diff(x,t,2)+(w0^2)*x==0

show(eqd)

#b)
desolve(eqd,x,ivar=t)

v0=var('v0')
x0=var('x0')

sol=desolve(eqd,x,ivar=t,ics=[0,x0,v0])

show(sol)
```

```python
#c) 
L=0.3924
g=9.81 #m/s^2
t=var('t')
w0=var('w0')
v0=var('v0')
x0=var('x0')
assume(w0,'integer')
assume(w0>0)
x0=0.15
v0=0
w0=sqrt(g/L)
R=sqrt(x0**2+(v0/w0)**2)
T=(2*pi/w0)
d=0
x=function('x')(t)
x=R*cos(w0*t-d)
show(x.full_simplify())
plot(x,t,-T,T)
```

```python
#Ex. 2
#a)
t=var('t')
lb=var('lb')
w0=var('w0')
x0=var('x0')
assume(w0,'integer')
assume(w0>0)
assume(lb^2>4*(w0**2))

x=function('x')(t)
deq1=diff(x,t,2)+lb*diff(x,t)+(w0^2)*x==0
f(t,w0,lb)=desolve(deq1,x,ivar=t)
show(f(t,w0,lb))
forget()
```
